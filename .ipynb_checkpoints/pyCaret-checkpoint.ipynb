## PyCaret Vs Scikit Learn Classfication 

```python
import pycaret 
import pandas as pd 
```

```python
irisData = pd.read_csv("IRIS.csv")
```

```python
seenData = irisData.sample(frac = 0.75)
 
unseenData = irisData.drop(seenData.index)
```

```python
irisData.columns
```

```python
from pycaret.classification import *
```

```python
exp_name = setup(data=seenData , target= "species")
```

```python
best= compare_models()
```

```python
newPred = predict_model(best, data=unseenData )
```

```python
evaluate_model(best)
```
